```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
style.use("ggplot")
%matplotlib inline
```

```python
data.head()
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(data)

centroids = kmeans.cluster_centers_
labels = kmeans.labels_

print(centroids)
print(labels)
```

```python
data['labels'] = labels
data
```

```python
data['labels'] = labels

#plt.plot(data, colors[data['labels'], markersize = 10)

group1 = data[data['labels']==0].plot( kind='scatter', x='x', y='y', color='DarkGreen', label="Group 1" )
group2 = data[data['labels']==1].plot( kind='scatter', x='x', y='y', color='Brown', ax=group1, label="Group 2" )
group3 = data[data['labels']==2].plot( kind='scatter', x='x', y='y', color='Blue', ax=group1, label="Group 3" )
group3 = data[data['labels']==3].plot( kind='scatter', x='x', y='y', color='red', ax=group1, label="Group 4" )






group1.legend(loc='upper center', bbox_to_anchor=(0.5, 1.05),
          ncol=3, fancybox=True, shadow=True)


plt.scatter(centroids[:, 0],centroids[:, 1], marker = "x", s=150, linewidths = 5, zorder = 10)

plt.show()
```
